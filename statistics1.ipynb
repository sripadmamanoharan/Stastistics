```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

#
```

```python
# define the dataset
data= [11,10,12,13,14,15,1008,14,14,26,88,97,10,92,35,76,87]
```

```python
# Z-score
outliers=[]

def detect_outliers(  data):
    threshold=3
    mean_1=np.mean(data)
    std_1=np.std(data)


    for y in data:
        z_score=(y-mean_1)/std_1
        if np.abs(z_score)>threshold:
            outliers.append(y)
    return outliers
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.hist(data)
plt.show()
```

```python
detect_outliers(data)
```
